```python
from control import dlqr
import numpy as np

A = np.array([[1,1],[0,1]])
B = np.array([[0],[1]])
y_max = 10
v_max = 5
f_max = 3
g_y = 1
g_v = 1
g_f = 1
Q = np.diag([g_y/(y_max**2),g_v/(v_max**2)])
R = np.array(g_f/(f_max**2))
out = dlqr(A,B,Q,R)
K = out[0]
S = out[1]
print("S=",S)
print("K=",K)
```
